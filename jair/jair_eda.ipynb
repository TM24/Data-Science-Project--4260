```python
# outlier checker using IQR 
def outliers(df, ft): 
    q3 = df[ft].quantile(0.25)
    q1 = df[ft].quantile(0.75)
    IQR = q3 - q1

    lower_bound = q1 - (1.5 * IQR)
    upper_bound = q3 + (1.5 * IQR)

    ls = df.index[ (df[ft] < lower_bound) | (df[ft] > upper_bound) ]
    return ls
```

```python
# add all things that quanitfy as an outlier to an array 
index = []
for feature in away: 
    index.extend(outliers(away, feature))
```
